```python
import geopandas as gpd
from shapely.geometry import Point
latitudes = []  
longitudes = []
```

```python
failed_stations_web = []
failed_stations_coord = []
```

```python
for station in stations:
    # URL of the webpage
    check_fill = 0
    url = f'https://www.ndbc.noaa.gov/station_page.php?station={station}'

    # Send an HTTP request to the webpage
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Get the HTML content
        html_content = response.text

    else:
        print(f"Failed to retrieve the webpage for station: {station}. Status code: {response.status_code}")
        failed_stations_web.append(station)
        break

    # Parse the HTML
    soup = BeautifulSoup(html_content, 'html.parser')
    # Find the station metadata section
    metadata = soup.find('div', id='stn_metadata')

    # Loop through all 'b' tags in the metadata section to find coordinates
    for bold_tag in metadata.find_all('b'):
        text = bold_tag.get_text()
        # Check if 'N' and 'W' are present in the text (indicating coordinates)
        if (' N ' in text and ' W ' in text) or (' S ' in text and ' W ' in text) or (' S ' in text and ' E ' in text) or (' N ' in text and ' E ' in text):
            # Split the text to extract latitude and longitude
            lat, lon = text.split()[0], text.split()[2]
            latitudes.append(lat)
            longitudes.append(lon)
            check_fill = 1

            # Print extracted latitude and longitude
            break  # Stop after finding the coordinates
    
    if check_fill == 0:
        print(f"Failed to retrieve the coordinates for station: {station}")
        failed_stations_coord.append(station)
```
